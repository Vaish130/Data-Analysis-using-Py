```python
#is there any missing values in data report
data.isnull().sum()

#sum of all values are not zero hence some data is missing like director,cast,country,dat_added,rating,duration
```

```python
#In column(like "director","cast","country") we are replace missing value with previous Row values
data["director"] = data["director"].fillna(method='ffill')
data["cast"] = data["cast"].fillna(method='ffill')
data["country"] = data["country"].fillna(method='ffill')

#set all other missing data
data.dropna(axis = 0, inplace=True)
```

```python
#content distribution of rating
rating = data.groupby(['rating']).size().reset_index(name='counts')
rating
```
